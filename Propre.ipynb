```python
import pandas
import time
import random
import pickle
import requests
import tqdm
```

```python
with open('countries.pickle','rb') as fp:
    countries = pickle.load(fp)
```

```python
filename = 'C:\\Users\\alexa\\Documents\\UTC\\IM04 - Depart etranger\\US02\\Data-sheet-STUDENTS.xlsx'
df = pandas.read_excel(filename, header = 2, usecols = 'B:AR')
```

```python
def get_indicator(country_code,indicator):
    head = 'http://api.worldbank.org/v2/country/'
    middle = '/indicator/'
    tail = '?format=json'
    url = head+country_code+middle+indicator+tail
    r = requests.get(url)
    try:
        data = r.json()
        for i in data[1]:
            if i['value'] is None:
                continue
            else:
                return i['value']
    except KeyError:
        return None
    except IndexError:
        return None
    except :
        return None
    
get_indicator('CHN','GC.TAX.IMPT.ZS')
    
```

```python
for i,v in df['Country (according to CN Gov):'].iteritems():
    country_name = v.split('(')[0].strip()
    country_code = countries[country_name]
    value = get_indicator(country_code,'GC.TAX.IMPT.ZS')
    if value is not None:
        df.at[i,'IMF WB\nCustoms and other import duties (% of tax revenue) 2016'] = value
    print(country_name,'done...')
    time.sleep(0.5)
df[['Country (according to CN Gov):','IMF WB\nCustoms and other import duties (% of tax revenue) 2016']]
```

```python
df_2 = pandas.read_excel('Data-sheet-STUDENTS.xlsx', header = 1)
links = {} #links and les points tels que GC.TAX.IMPT.ZS 
for i,v in df_2.loc()[0].iteritems():
    if 'data.worldbank' in i:
        links[v] = i.split('/')[-1]
```

```python
for i,v in df['Country (according to CN Gov):'].iteritems():
    country_name = v.split('(')[0].strip()
    country_code = countries[country_name]
    for each in links:
        value = get_indicator(country_code,links[each])
        time.sleep(0.02)
        if value is not None:
            df.at[i,each] = value
        print(each,'done...')
    print(country_name,'done...')
```
